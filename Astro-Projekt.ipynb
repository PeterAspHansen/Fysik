## **Introduktion**

> Den nuværende standard model for universets udvikling bygger på 'Hot Big Bang' modellen. Denne model beskriver, hvordan universet har udvidet sig fra et varmt og tæt stadie til det nuværrende relativt koldere og mindre kompakte stadie. Selvom denne teori er understøttet af universets udvidelse i dag, opstår der stadig nogle problemer ved det tidlige univers, da det var strålingsdomineret. Disse problemer er kendt som fladhedsproblemet, horisontproblemet og monopolproblemet. Dette projekt tager udgangspunkt i fladhedsproblemet, men det er vores intention også kort at forklare de to andre problemer. De er forklaret nedenfor: 

- Horisontproblemet. 

Når vi betragter forskellige områder i universet, så vil alle punkter på en kosmisk skala have samme temperatur på 2.7 K, på trods af at alle punkter ikke er kausalt forbundne. Det er meget usandsynligt, at to punkter som ikke har udveklset information, og dermed ikke er causalt forbundne, har den samme temperatur, og det er dermed endnu mere usandsynligt, at alle punkter har samme temperatur. Vi har dermed et problem i teorien, som (spoiler) kan forklares vha. inflation. 

- Monopolproblemet

Det opstår, en forskel mellem teorien og det vi observerer, når vi betragter magnetiske monopoler i universet. Der er mangel på magnetiske monopole i universet, som i teorien burde have været opstået, da universet var varmt og tungt. 

## Fladhedsproblemet
Fladhedsproblemet omhandler krumningen af rum-tiden som er afhængig af tæthedsparametrene jf. ved Friedmann-ligningen hvor $(1-Ω)$ repræsenterer den kritiske værdi. I et flat univers er den kritiske værdi meget tæt på 0. Vi har derfor etableret $Ω_0$ i dag er meget tæt 1 og det viser sig at $Ω_0$ må have være endnu tættere på 1 i det tidlige univers. Havde $Ω_0$ ikke være utroligt tæt på 1 ved det tidlige univers, ville universet formegentlig enten forlængst have kollapset (Big Crunch) eller udvidet sig betydeligt mere, end det vi observerer i dag (Big Bore). Begge tilfælde ville have haft katastrofale konsekvenser for vores eksistens. I det første tilfælde ville universets levetid have været betydeligt mindre, og som resultat af dette ville blandt andet stjerner ikke have haft tid til at blive dannet. I det andet tilfælde ville universet have udvidet sig så hurtigt, at stof ikke ville kunne samles og danne stjerne. Vi burde derfor kalde det fladhedsmiraklet fremfor fladshedsproblemet. Grunden til at vi kalder det et problem skyldes, at ved "Hot Big Bang" modellen, kan tæthedsparameterens værdi kun tilskrives som held. I teorien ville $Ω_0$ kunne være $10^6$ eller $10^{-6}$, uden at nogen fysiske love er brudt. Hvis værdierne for tæthedsparametrene havde været en smule anderledes i det tidlige univers, ville universets udvikling have set fuldstændig anderledes ud.  

Så hvorfor er vi så heldige, at tæthedsparameteren er så tæt på 1? 
Svaret er sandsynligvis, at der er en mekanisme, som driver $Ω_0$ mod 1. Denne mekanisme kommer vi til at stifte bekendskab med senere, men du kan sikkert allerede gætte, hvad den hedder.

Vi vil nu beskrive, hvordan universets krumning eller mangel deraf, har set ud fra dets begyndelse til i dag. Dette gør vi ved hjælp af Friedmannligningen.

Friedmannligningen kan opskrives som udtrykt ved:

$\Omega(a)-1=\frac{\kappa c^2}{H^2a^2R^2}$

Mens at hvis vi sætter skalarfaktoren til 0, dvs. bestemmer tæthedsparametren til nutiden ($\Omega_0$), så finder vi at:

$\Omega_0-1=\frac{\kappa c^2}{H_0^2R^2}$

Vi kan her isolere: $\frac{\kappa c^2}{R}$ og sætte denne ind i ovenstående udtryk for Friemannlignignen:

$\Omega(a)-1=\frac{H_0^2(\Omega_0-1)}{H^2a^2}$

I vores univers forventer vi at den generelle Freidmanlignignen desuden kan beskrives ved de tre tæthedsparametre samt krumningen i udtrykket:

$H^2=H_0^2(\frac{\Omega_{r,0}}{a^4}+\frac{\Omega_{m,0}}{a^3}+\Omega_\Lambda+\frac{1-\Omega_0}{a^2})$

Hvis vi nu kun betragter et strålingsdomineret univers kan vi sætte værdierne for tæthedsparametrene for stof og mørk energi til 0. I dette scenarie vil den samlede tæthedsparameter $\Omega_0$ være udelukkende givet ved $\Omega_{r,0}$. Vi har altså at:

$H^2=H_0^2(\frac{\Omega_{0}}{a^4}+\frac{1-\Omega_0}{a^2})$

Hvis vi sætter dette ind i udtrykket for krumningen af universet får vi altså:

$\Omega(a)-1=\frac{(\Omega_0-1)a^4}{a^2\Omega_0+a^4(1-\Omega_0)}$

$\Omega(a)=\frac{(\Omega_0-1)a^2}{\Omega_0+a^2(1-\Omega_0)}+1$

$\Omega(a)=\frac{(\Omega_0-1)a^2+\Omega_0+a^2(1-\Omega_0)}{\Omega_0+a^2(1-\Omega_0)}$

$\Omega(a)=\frac{\Omega_0}{\Omega_0+a^2(1-\Omega_0)}$


Ved samme fremgangsmåde kan man finde den samlede tæthedsparameter som funktion af skalarfaktoren for et univers domineret af stof, hvor $H^2=H_0^2(\frac{\Omega_{0}}{a^3}+\frac{1-\Omega_0}{a^2})$:

$\Omega(a)=\frac{\Omega_0}{\Omega+(1-\Omega_0)a)}$

Og for et univers domineret af mørk energi, hvor $H^2=H_0^2(\Omega_0+\frac{1-\Omega_0}{a^2})$:

$\Omega(a)=\frac{a^2\Omega_0}{(a^2-1)\Omega_0+1}$

I nedenstående har vi plottet disse tre udtryk fundet for de tre tæthedsparametre.

## Interaktive grafer
 På de nedenstående interaktive grafer, kan du selv prøve at ændre $Ω_0$, som styrer krumningen i dag, til $a=1$. Derefter kan du se grafer for $Ω_0=0.98$ og $Ω_0=1.02$. Overvej hvad dette betyder for universets krumning i løbet af dets levetid alt efter om det har været domineret af stråling, stof eller mørk energi.

```python
#Stråling slider
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

x = np.arange(0, 3, .001)

def plot_func(omega_0):
    plt.xlim(0, 3)
    plt.ylim(0, 3)
    plt.title('Krumning i strålingsdomineret univers')
    plt.xlabel('a')
    plt.ylabel('omega(a)')
    y = omega_0/(omega_0+x**2*(1-omega_0))
    plt.plot(x, y)

interact(plot_func, omega_0 = widgets.FloatSlider(value=1,
                                               min=0,
                                               max=1.1,
                                               step=0.01))
```

```python
#Stråling graf
from numpy import *
import math
import matplotlib.pyplot as plt


a = linspace(0, 10, 400)
omega098=0.98/(0.98+a**2*(1-0.98))
omega102=1.02/(1.02+a**2*(1-1.02))


plt.xscale('log')
plt.plot(np.log(a), omega098, 'r',label='Ω_0=0.98') 
plt.plot(np.log(a), omega102, 'b',label='Ω_0=1.02') 
plt.xlim(0.1, 1.9)
plt.ylim(0.7, 1.5)
plt.xlabel('log(a)')
plt.ylabel('omega(a)')
plt.title('Strålingsdomineret univers')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)



plt.show()
```

## Hvordan foregik inflationen:

Inflation er en kort periode i det meget tidlige univers, hvor universet ekspanderede eksponentielt. Dette skete omkring 10^{-36} s efter big bang. Vi vil se på hvordan, dette løser vores problem senere.
Men hvordan kan sådan et fænomen overhovedet ske, og hvad bestemte, hvornår det startede, og hvornår det sluttede?

Der er ikke pt. konsensus om hvilken teori, som beskriver det bedst, men her tages der udgangspunkt i teorien fra i Introduction to Cosmology af Barbara Ryden.
Denne teori antager, at universet indeholder skalarfeltet ϕ(r,t) , hvilket minder om det, vi kender fra Higgsfeltet i dag. Dette felt kan have været Higgsfeltet, men det er stadig ikke klart, derfor bliver der typisk omtalt som inflatonfeltet. Feltet er altså en funktion af position og tid. Dette felt vil have en potentiel energi associeret med sig, som kaldes V.

Skalarfeltet har et globalt minimum, når potentiallet er 0, men antag at feltet starter ved $\Phi$ og den potentielle energi $V_0$. Her befinder feltet sig i “et falsk vakuum”. Hvis dette felt bliver skubbet fra $\Phi=0$ til $\Phi>0$, vil det begynde at gå med det ægte vakuum, da plateauet er meget svagt skrånende, som altså er ved 0.
Idet inflatonfeltet går fra det falske vakuum til det ægte, giver det anledning til en udvidelse der kaldes inflationen. Dette kan visualiseres med følgende graf:

62EB39F6-65CF-4564-84FB-A399779280E4.jpeg

Denne langsomme ændring i potentialet kan give anledning til en kosmologisk konstant, som kan drive en eksponentiel forstørrelse af skalafaktoren idet: $a(t)=a(t_i)e^{H_i(t_f-t_i)} $
Denne eksponentielle forstørrelse af skalafaktoren vil resultere i en eksponentiel udvidelse af universet, hvilket netop er inflation.

Når inflatonfeltet endelig når den sande vakuumtilstand, vil det oscillere mellem lavere og lavere energipotentialer, indtil at det når at afgive dens energi igennem noget kaldet “Higgs friktion”. Denne sidste afgivelse af energi har store konsekvenser for universets indhold. Under inflationen ville koncentrationen af masse og stråling nemlig blive ekstremt forringet. Universet ville dermed have været meget koldere og mere tomt end vi ser det i dag. Afgivelsen af energi under “Higgs friktionen” løser dog dette problem i form en slags genopvarmning af universet til temperaturer, som det forventes at universet havde i det meget tidlige univers.

## En løsning på fladhedsproblemet:

Men hvordan løser inflationen så fladhedsproblemet, dvs. hvordan kan det være at universets tæthedsprarameter $\Omega_0$ var så tæt på én. Dette spørgsmål vil følgende video gennemgå. Her husker vi på, at under inflationen var universet kosmiskkonstant domineret, og tæthedsparametren var derfor beskrevet ved: $\Omega(a)=\frac{a^2\Omega_0}{(a^2-1)\Omega_0+1}$.

Som beskrevet tidligere, vokser a hurtigt, idet den vokser eksponentielt som beskrevet af ligningen 

$a(t)=a(t_i)e^{H_i(t_f-t_i)}$

Hvis tidsintervallet for inflation er langt i forhold til Hubble tiden $H_i^{-1}$, vil universet vokse voldsomt hurtigt, og som beskrevet i videoen vil universet dermed også blive flat. 


Inflationen løser også horisontproblemet. I det at universet har været så tæt forbundet og senere blevet skildt igennem inflationen, så løser det netop problemet ved at alle punkter i universet har samme temperatur på kosmisk skala. Alle punkter var derfor før inflationen kausalt forbundende, og kunne derigennem opnå samme temperatur.

Monopolproblemet løses også ved inflationen i det at koncentrationen af monopoler vil blive så difus under den eksponentielle udvidelse af universet at det vil være meget usandsynligt nogensinde at detektere dem.
